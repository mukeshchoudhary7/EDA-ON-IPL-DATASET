# Who is the most consistent batsman and bowler in each team and also in the entire league?

```python
#batter_batsman in ipl most run
batsman_by_player = ipl.groupby('batter')['total_run'].sum()
batsman_by_player.sort_values(ascending=False).head(5)
```
